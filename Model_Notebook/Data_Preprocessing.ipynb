# Preprocessing the Dataset after downloading from Kaggle

```python
import numpy as np
import pandas as pd
```

#### loading the dataset (downloaded from kaggle)


```python
df = pd.read_csv("US Zip Codes from 2013 Government Data.csv")
df.shape
```

```python
df.head(5)
```

#### Removing duplicate ZIP rows

```python
df = df.drop_duplicates(subset=["ZIP"])
df.shape
```

```python
# Use only valid rows
df = df.dropna(subset=['LAT', 'LNG'])
print(df.shape)
```

```python
# remove rows with null values
df = df.dropna()
print(df.shape)
```

#### keeping only mainland USA coordinates

```python
df = df[
    (df["LAT"] >= 24) & (df["LAT"] <= 50) &
    (df["LNG"] >= -125) & (df["LNG"] <= -66)
]
df.shape
```

```python
#saving the final dataset (Used in the backend)
df.to_csv("Zip_codes.csv", index=False)
```
